```python
import pandas as pd
import numpy as np
```

#Importing California dataframe

```python
from sklearn.datasets import fetch_california_housing
```

```python
#storing the dataset in a variable
california_dataset=fetch_california_housing()
print(california_dataset)
```

```python
#type of the dataset
type(california_dataset)
```

#Creating pandas DataFrame

```python
california_df=pd.DataFrame(california_dataset.data,columns=california_dataset.feature_names)
#displaying the first 5 rows of the dataframe
california_df.head()
```

```python
#shape of the dataframe
california_df.shape
#type of the dataframe
type(california_df)
```
